```python
# Criação das variaveis para representar a população, a amostra e o valor de k

populacao = 150
amostra = 15
k = ceil(populacao / amostra)
print(k)
```

```python
# Definicao do valor randômico para inicializar a amostra, iniciando em 1 até k + 1

r = np.random.randint(low = 1, high = k + 1, size = 1)
print(r)
```

```python
# Criamos um for para somas os próximos valores, basseado no primeiro valor r que foi definido acima

acumulador = r[0]
sorteados = []
for i in range(amostra):
    #Print(acumulador)
    sorteados.append(acumulador)
    acumulador += k
print(sorteados)

```
